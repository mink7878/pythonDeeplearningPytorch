## 2. 문자를 숫자로 표현하는 방법

- NLP에서 input은 문자
- 하지만 딥러닝 모델의 input은 숫자 데이터  
=> 그래서 모델 학습을 위해 가장 먼저 해야 할 일은 주어진 __문자를 숫자로 표현__하는 일!


- (참고) 앞으로 데이터의 기본 형태는 __문장(sentence)으로 가정__ (대부분 딥러닝 모델의 input을 문장으로 사용하고 있기 때문)

- 인간이 문장이나 글을 어떤 과정을 거쳐 이해하고 있을까?  
> 1) 문장을 __의미 있는 부분(단위)__으로 나눈다; __<span style="color:blue">Text Segmentation</span>__  
> 2) 나눠진 의미 있는 부분을 숫자로 바꿔 __문장을 숫자로 표현__한다; __<span style="color:blue">Representation</span>__

- 1) 과정에서 쓸 수 있는 가장 간단한 방법은 __띄어쓰기__를 이용하는 것.

```python
S1 = '나는 책상 위에 사과를 먹었다'
S2 = '알고 보니 그 사과는 Jason 것이었다'
S3 = '그래서 Jason에게 사과를 했다'
```

```python
S1.split()
```

```python
S2.split()
```

```python
S3.split()
```

- 위와 같이 문장을 의미 있는 부분으로 나누는 과정을 __tokenization__이라 하며, 
- 쪼개진 부분을 __token__이라 부른다.
- 컴퓨터에게 '여기서부터 여기까지의 문자의 나열은 특정 의미가 있는 부분이니 하나로 인식하렴.'이라고 가르쳐주는 첫 과정

```python
list(S1)
```

- 위 코드처럼 __각 글자로 분리__해 문장을 이해하는 것보다는 __띄어쓰기__를 통한 단어들의 나열로 이해하는 것이 훨씬 쉽다.

```python
token2idx = {} # token을 저장해 놓은 사전
index = 0

for sentence in [S1, S2, S3]:
    tokens = sentence.split()
    for token in tokens:
        if token2idx.get(token) == None: # get(x) : x라는 key에 대응되는 value 
            token2idx[token] = index
            index += 1
            
print(token2idx)
```

```python
def indexed_sentence(sentence):
    return [token2idx[token] for token in sentence]
```

```python
S1_i = indexed_sentence(S1.split())
print(S1.split())
print(S1_i)
```

```python
S2_i = indexed_sentence(S2.split())
print(S2.split())
print(S2_i)
```

```python
S3_i = indexed_sentence(S3.split())
print(S3.split())
print(S3_i)
```

- 그런데 위와 같은 방식으로 문자를 숫자로 바꿔 모델에게 주고, 알아서 패턴을 찾으라고 하면 모델이 학습을 잘 할까?
- 다음과 같은 의구심이 들 것이다.  
> - '배'나 '복숭아'처럼 __내가 만든 사전에 없는 단어__가 나오면 어떻게 하지?
> - 'Jason'이나 'Jason에게'는 사실상 __같은 단어__ 같은데 조사 차이로 __굳이 따로 나눠야 하나__?  
> - S1의 '사과'가 S3의 '사과'가 아닌데, __같은 '사과'__로 처리하네?

### 2.1 Corpus & Out-Of Vocabulary(OOV)

```python
'''
S4 = '나는 책상 위에 배를 먹었다'

indexed_sentence(S4.split())
'''
# KeyError: '배를'
```

- 인간이 보기엔 '나는 책상 위에 __사과를__ 먹었다'와 '나는 책상 위에 __배를__ 먹었다'가 비슷하다고 느낄 수 있다.
- 그러나 컴퓨터의 입장은 다르다. 
- token을 저장해뒀던 __vocabulary에는 없는 '배를'이라는 token__을 찾을 수 없기 때문
- 이렇게 token을 저장해둔 vocabulary에 token이 없어서 처음 본 token이 나오는 현상  
    => __Out-Of Vocabulary(OOV)__
- 보통 이런 상황을 대비해 __특수한 token인 \<unk> token__ (unknown의 앞 글자를 뜻함)을 만들어 vocabulary에 없는 token이 나올 경우 "\<unk>"로 변환하도록 처리한다.

```python
token2idx
```

```python
# 기존 token 사전에 <unk> token 추가

token2idx = {t: i+1 for t, i in token2idx.items()}
token2idx['<unk>'] = 0
token2idx
```

```python
# token이 없을 경우 <unk> token의 0을 치환

def indexed_sentence_unk(sentence):
    return [token2idx.get(token, token2idx['<unk>']) for token in sentence] # token없을 경우 token2idx['<unk>'] = 0
```

```python
S4 = '나는 책상 위에 배를 먹었다'

indexed_sentence_unk(S4.split())
```

- 하지만 이렇게 새로 나온 단어마다 \<unk> token으로 처리하는 것보다 __애초에 사전을 풍부하게 만들면 OOV 문제가 해결__될 것.
- 그러러면 다양한 token을 많이 모을 수 있는 __문장__이 필요
- 더 많은 문장으로 S4, S5, ...를 가져와 token을 모은다면 vocabulary가 풍부해지고 OOV 문제도 일어날 확률이 줄어들 것이다.
- 그뿐 아니라 비슷한 영역의 문장만 계속 반복해 등장한다면 OOV의 문제가 나타날 위험성이 있을 것이므로 __corpus domain도 다양할수록 좋을 것.__
- 그래서 유명한 몇 가지의 open corpus가 공개돼 있음.
- 큰 기업의 연구에서는 corpus를 잘 만드는 것도 모델 성능 향상의 큰 요인이라 생각해 질 좋은 corpus를 만들어 공개하기도 함.

- 그런데 corpus가 커질수록 사전의 크기 역시 커질 것.
- 그리고 사전의 크기가 커질수록 우리가 만드는 모델의 사이즈도 커진다.  
    => 너무 커질 경우 메모리에 부담 줄 수 있기 때문에 __token을 효율적으로 만드는 방법__ 자연스럽게 고민
- 그뿐 아니라 S2의 "Jason" token과 S3의 "Json에게" token과 같은 부분을 __다른 token으로 나누는 것은 효율적인 방법이 아닐 것!__
- 효율적인 token 사전을 만들려면 앞서 사용했던 __띄어쓰기보다 좋은 tokenizer가 필요__함.

### 2.2.1 Character based tokenization

- 띄어쓰기가 아니라 __<span style="color:blue">글자(character)</span>__를 token으로 사용하면 어떨까?  
> 1. 우선, token 사전의 크기가 줄어든다.  
- 특수 문자나 대문자를 고민한다 해도 기존의 띄어쓰기를 통한 tokenization보다 사전의 크기 면에서 부담이 되지 않을 양.  
> 2. OOV 현상을 사실상 없앨 수 있다.  
- __글자가 사실상의 최소 단위__이므로 자음과 모음으로만 나타내는 한국어의 단축어가 아닌 이상 모든 글자를 표현할 수 있기 때문
- 하지만 글자를 이용하면 __어떤 단어나 문장도 표현할 수 있기 때문__에 인터넷에서 만들어지는 단축어와 신조어의 생성 속도도 걱정할 필요 x

----------------------------------------------------------------------------------------------------------------

- 사전에 넣을 문자부터 정리해보자.

```python
# 영어 unicode

print([chr(k) for k in range(65, 91)]) # 대문자
print([chr(k) for k in range(97, 123)]) # 소문자
```

```python
# 특수 문자 및 숫자 unicode

print([chr(k) for k in range(32, 48)])
print([chr(k) for k in range(58, 65)])
print([chr(k) for k in range(91, 97)])
print([chr(k) for k in range(123, 127)])
print([chr(k) for k in range(48, 58)])
```

```python
# 한국어 unicode
print([chr(k) for k in range(int('0xAC00', 16), int('0xD7A3', 16) + 1)]) # 모든 완성형 한글 11172자
```

```python
print([chr(k) for k in range(int('0x3131', 16), int('0x3163', 16) + 1)]) # 자모
```

-----------------------------------------------------------------------------------

- 이제 이를 바탕으로 character token vocabulary를 만들어보자.

```python
idx2chr = {0 : '<pad>', 1 : '<unk>'}

srt_idx = len(idx2chr)
for x in range(32, 127):
    idx2chr.update({srt_idx : chr(x)})
    srt_idx += 1
```

```python
print(idx2chr)
```

```python
# 한글 추가

for x in range(int('0x3131', 16), int('0x3163', 16) + 1):
    idx2chr.update({srt_idx : chr(x)})
    srt_idx += 1
    
for x in range(int('0xAC00', 16), int('0xD7A3', 16) + 1):
    idx2chr.update({srt_idx : chr(x)})
    srt_idx += 1
```

```python
print(idx2chr)
```

```python
chr2idx = {v:k for k, v in idx2chr.items()}
print(chr2idx)
```

- 기존에 있었던 문장은 물론, 줄임말('ㅇㅋ! ㄱㅅㄱㅅ')이나 신조어에 대한 OOV를 걱정할 필요 x
- 세상에 있는 모든 단어를 저장하는 것보다는 모든 글자 수를 등록하는 것이 훨씬 작은 크기를 갖기 때문

- 그러면 좋은 점만 있을까?
- 실제 최근 딥러닝 모델에서 사용하는 token은 대부분 글자 단위가 x
- 글자 단위의 tokenization의 문제 중 하나는 __표현법에 대한 학습이 어렵다__는 것이다.
- 글자 하나는 보통 특정 의미를 갖고 있지 않다.
- 학습의 부담을 줄이기 위해 __글자보다 의미를 가진 단위__이자 기존의 __띄어쓰기보다 효과적인 단위__를 찾아내는 방법 연구

### 2.2.2 n-gram Tokenization

- __token이 글자라면__ OOV 현상이 벌어지는 일은 거의 없을 것.
- 하지만 __글자 하나에는 의미가 거의 없기 때문에__ 모델의 구조를 만들 때 __글자의 특정 연속성이 의미를 가진 단어라는 것을 학습하게__ 만들고 그것을 문장이나 문단과 같은 더 긴 형태의 글을 이해하도록 만들어야 한다.  
    => 절대 쉬운 문제 x
- 짧은 문장이나 단어 수준의 이해를 목적으로 가진 task가 아닌, 문단의 수준 이해나 문장의 생성에 대한 task에서는 글자 기반의 tokenizer를 사용하는 것은 어려울 수 o
- 그래서 __글자보다는 좀 더 긴 형태의 token을 만들어내기 위해 사용하는 방법 중 한 가지가 <span style="color:blue">n-gram</span>!__

__<span style="color:blue">* n-gram 방법 *</span>__
- 1개 단위가 아닌 여러 개(n개)의 연속된 __윈도우를 단위로__ 살펴보기 위해 나온 개념.
- 글자 기준의 uni-gram(n=1)이 앞절의 character based tokenization이라 볼 수 o

```python
S1 = '나는 책상 위에 사과를 먹었다'

print([S1[i:i+1] for i in range(len(S1))]) # uni-gram
```

```python
print([S1[i:i+2] for i in range(len(S1))]) # bi-gram
```

```python
print([S1[i:i+3] for i in range(len(S1))]) # tri-gram
```

- 각 글자를 기준으로 윈도우가 이동하면서 token으로 잡아내는 것을 알 수 o
- uni-gram에서는 의미 없는 글자가 token이었다면 bi-gram과 tri-gram으로는 띄어쓰기나 글자로는 잡을 수 없었던 '책상', '사과', '먹었다' 등의 token을 찾을 수 있고, 이런 __의미를 가진 token__은 모델 학습을 도와줄 수 있을 것!
- 글자 기준이 아니라 __띄어쓰기 기준의 token__에 대해서도 n-gram을 사용할 수 있다.
- 특히 영어의 경우 띄어쓰기가 단어의 기준이므로 연속적으로 사용되는 용어를 잘 찾아낼 수 o

- uni-gram, bi-gram, tri-gram의 token을 모아 uni-gram보다 좋은 결과를 만들어낼 수 있겠지만, 단점도 있다.
> 1. 쓸모 없는 조합이 너무 많이 생김
> 2. token 사전이 과하게 커짐  
- 그래서 __n-gram의 이점을 챙기면서 그중 의미가 있는 것들만 token으로 사용하는 방법이 <span style="color:blue">BPE!</span>__

### 2.2.3 BPE(Byte Pair Encoding)

- data compression 분야에서 사용됐던 개념
- __반복적으로 나오는 데이터의 연속된 패턴을 치환__하는 방식을 사용해 __데이터를 좀 더 효율적으로 저장__하는 개념
- 자주 나오는 패턴이 있다면 다른 변수로 치환해 표현하자는 것.
- 치환했던 부분만 추가로 기억하고 있으면 됨.

```python
'''
abbcabcab

if X = ab, 
=> XbcXcX

if Y = cX,
=> XbYY
'''
```

- 이 개념을 tokenization에 적용해 소개한 논문이 __\<Neural Machine Translation of Rare Words with Subword Units> (Sennrich 등, 2015)__
- 이 논문을 시작으로 BPE는 NLP에서 자주 사용하는 tokenizer algorithm이 됨.
- n-gram에서 쓸모없이 많아지는 연속된 글자의 나열이 아닌 __'여러 번 나타나는 글자의 나열'은 의미가 있다고 생각__해 따로 token을 만드는 방식으로 개선했다고 볼 수 o
- 단어를 __'자주 나오는 글자의 나열'이라는 subword로 한 번 더 나눠 표현__함으로써 __더욱 효율적인 token__을 만들 수 있다.

```python
# 알고리즘 1 : Learn BPE operations

import re, collections

def get_stats(vocab):
    pairs = collections.defaultdict(int) # defaultdict() : 키의 개수 셀 때 사용
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols) - 1):
            pairs[symbols[i], symbols[i+1]] += freq
    return pairs
```

```python
# (3) 두 글자를 합쳐 기존의 사전의 단어의 수 수정
def merge_vocab(pair, v_in):
    v_out = {}
    bigram = re.escape(' '.join(pair)) # 특수문자 escape 처리
    p = re.compile(r'(?<!\\S)' + bigram + r'(?<!\\S)') # re.compile(검색할문자열) 
    for word in v_in:
        w_out = p.sub(''.join(pair), word) # re.sub('패턴', '바꿀문자열')
        v_out[w_out] = v_in[word]
    return v_out
```

```python
# (1) 단어의 '횟수' 기록한 사전 만들기 (띄어쓰기 기반의 tokenization)
# 이때 사전의 단어 글자는 모두 띄어 표현

vocab = {'l o w </w>' : 5, 
        'l o w e r </w>' : 2,
        'n e w e s t </w>' : 6, 
        'w i d e s t </w>' : 3}
```

```python
# (2) 각 단어에 대해 연속된 2개의 글자의 숫자를 세어 가장 많이 나오는 글자 2개의 조합을 찾음.
# (character bi-gram)

get_stats(vocab)
```

```python
num_merges = 10 # 미리 정해 놓은 횟수만큼 (2)~(3)번 반복

for i in range(num_merges): # 10번 반복
    pairs = get_stats(vocab) # (2)
    best = max(pairs, key=pairs.get) # (2)
    # (2) 각 단어에 대해 연속된 2개의 글자의 숫자를 세어 가장 많이 나오는 글자 2개의 조합을 찾음.
    # character bi-gram
    
    vocab = merge_vocab(best, vocab) # (3)
    
    print(f'Step {i+1}')
    print(best)
    print(vocab)
    print('\n')
```

- 결과를 보면 'est', 'low' 등 글자의 연속인 subwords를 찾을 수 있음.
- 이를 이용하면 token을 character uni-gram이나 bi-gram을 섞어 쓰는 방식보다 효율적으로 구성할 수 o

- 이전의 띄어쓰기나 character n-gram tokenization 결과와 비교해보기 위해 같은 문장 (S1, S2, S3)을 BPE로 다시 tokenization해보면서 어떤 차이가 있는지 확인해보자.

```python
S1 = '나는 책상 위에 사과를 먹었다'
S2 = '알고 보니 그 사과는 Jason 것이었다'
S3 = '그래서 Jason에게 사과를 했다'

token_counts = {}
index = 0

for sentence in [S1, S2, S3]:
    tokens = sentence.split()
    for token in tokens:
        if token_counts.get(token) == None:
            token_counts[token] = 1
        else:
            token_counts[token] += 1

print(token_counts)
```

```python
token_counts = {" ".join(token):counts for token, counts in token_counts.items()}
print(token_counts)
```

```python
num_merges = 10

for i in range(num_merges):
    pairs = get_stats(token_counts)
    best = max(pairs, key=pairs.get)
    token_counts = merge_vocab(best, token_counts)
    print(f'Step {i+1}')
    print(best)
    print(token_counts)
    print('\n')
```

- 여러 번의 병합 과정을 거치면서 띄어쓰기 token과 거의 비슷한 token이 생성됨.
- 하지만 __띄어쓰기에서는 잡아낼 수 없었던 '사과'__를 token으로 따로 뽑아낸 것을 확인할 수 o
- 또한, __uni-gram이나 bi-gram으로 잡을 수 없었던 'Jason'__ token도 잡아낸 것을 확인할 수 o
- 이전에 '사과를'을 '배를'로 바꿀 경우에 나타나는 OOV는 BPE에서도 나타나겠지만, 이는 corpus가 늘어나면 해결할 수 있는 문제임.
- 이처럼 __여러 가지 방법의 장점을 잘 섞어 놓은 <span style="color:blue">BPE 알고리즘</span>__은 최근 NLP의 대표적인 tokenizer로 자리매김함.

- 위 결과물을 자세히 보면 마지막 step 10의 결과물이 더 좋지 않다고 생각할 수 있음.  
    => '사과'나 'est' 등은 분리해 놓는 중간 단계가 더 좋은 성능을 지니고 있다고 볼 수 있기 때문.
- 하지만 위 내용은 설명을 위해 상당히 적은 양의 corpus에서 학습한 경우이므로 마지막 스텝이 무조건 완벽한 답변이라 보기는 어렵.
- 더 큰 corpus에서 적절한 iteration을 지정해 만들면 위 결과보다 더 적합한 vocabulary를 만들 수 있음.
